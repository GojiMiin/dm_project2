```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
```

```python
readIn = pd.read_csv('./bank/bank.csv')
```

```python
raw = readIn.iloc[:,0]
```

```python
age = []
job = []
marital = []
edu = []
default = []
balance = []
housing = []
loan = []
contact = []
day = []
month = []
duration = []
campaign = []
pdays = []
previous = []
poutcome = []
y = []
```

```python
for i in raw:
    lineArr = i.split(";")
    
    age.append(int(lineArr[0]))
    job.append(lineArr[1].replace('"', ''))
    marital.append(lineArr[2].replace('"', ''))
    edu.append(lineArr[3].replace('"', ''))
    default.append(lineArr[4].replace('"', ''))
    balance.append(int(lineArr[5]))
    housing.append(lineArr[6].replace('"', ''))
    loan.append(lineArr[7].replace('"', ''))
    contact.append(lineArr[8].replace('"', ''))
    day.append(int(lineArr[9]))
    month.append(lineArr[10].replace('"', ''))
    duration.append(int(lineArr[11]))
    campaign.append(int(lineArr[12]))
    pdays.append(int(lineArr[13]))
    previous.append(int(lineArr[14]))
    poutcome.append(lineArr[15].replace('"', ''))
    y.append(lineArr[16].replace('"', ''))
```

```python
zippedList =  list(zip(age, job, marital, edu, default, balance, housing, loan, contact, day, month, duration, campaign, pdays, previous, poutcome, y))
```

```python
new_pd = pd.DataFrame(zippedList, columns = ['age' , 'job', 'marital', 'edu', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'y'])
```

```python
balance_row = new_pd.iloc[:,5]
duration_row = new_pd.iloc[:,11]
```

```python
def iqrmethod(row):
    quantile1, quantile3 = np.percentile(row,[25,75]) #25,75 percentile
    ## Find the IQR
    iqr_value = quantile3 - quantile1
    ## Find the lower bound value and the higher bound value
    lower_bound_val = quantile1 - (1.5 * iqr_value) 
    upper_bound_val = quantile3 + (1.5 * iqr_value)
    return lower_bound_val,upper_bound_val
```

```python
lower_bound_balance,upper_bound_balance = iqrmethod(balance_row)
lower_bound_duration,upper_bound_duration = iqrmethod(duration_row)
```

```python
new_pd_filter_bal = new_pd.loc[(new_pd["balance"] > lower_bound_balance) & (new_pd["balance"] < upper_bound_balance) & (new_pd["balance"] > 0)]
new_pd_filter_bal
```

```python
new_pd_filter_bal_duration = new_pd_filter_bal[(new_pd_filter_bal["duration"] > lower_bound_duration) & (new_pd_filter_bal["duration"] < upper_bound_duration) & (new_pd["duration"] > 0)]
new_pd_filter_bal_duration
```

```python
cols_to_norm = ['balance','duration']
new_pd_filter_bal_duration[cols_to_norm] = new_pd_filter_bal_duration[cols_to_norm].apply(lambda x: (x - x.min()) / (x.max() - x.min()))
```
