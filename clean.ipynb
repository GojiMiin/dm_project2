```python
import pandas as pd
import numpy as np
```

```python
readIn = pd.read_csv('./bank/bank.csv')
```

```python
raw = readIn.iloc[:,0]
```

```python
age = []
job = []
marital = []
edu = []
default = []
balance = []
housing = []
loan = []
contact = []
day = []
month = []
duration = []
campaign = []
pdays = []
previous = []
poutcome = []
y = []
```

```python
for i in raw:
    lineArr = i.split(";")
    
    age.append(int(lineArr[0]))
    job.append(lineArr[1].replace('"', ''))
    marital.append(lineArr[2].replace('"', ''))
    edu.append(lineArr[3].replace('"', ''))
    default.append(lineArr[4].replace('"', ''))
    balance.append(int(lineArr[5]))
    housing.append(lineArr[6].replace('"', ''))
    loan.append(lineArr[7].replace('"', ''))
    contact.append(lineArr[8].replace('"', ''))
    day.append(int(lineArr[9]))
    month.append(lineArr[10].replace('"', ''))
    duration.append(int(lineArr[11]))
    campaign.append(int(lineArr[12]))
    pdays.append(int(lineArr[13]))
    previous.append(int(lineArr[14]))
    poutcome.append(lineArr[15].replace('"', ''))
    y.append(lineArr[16].replace('"', ''))
```

```python
zippedList =  list(zip(age, job, marital, edu, default, balance, housing, loan, contact, day, month, duration, campaign, pdays, previous, poutcome, y))
```

```python
new_pd = pd.DataFrame(zippedList, columns = ['age' , 'job', 'marital', 'edu', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome', 'y'])
```
